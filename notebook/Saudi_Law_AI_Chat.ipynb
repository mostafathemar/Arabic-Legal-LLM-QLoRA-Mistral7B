```python
base_model = "mistralai/Mistral-7B-Instruct-v0.2"

tokenizer = AutoTokenizer.from_pretrained(base_model)
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    device_map="auto",
    torch_dtype=torch.float16
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=200,
    temperature=0.7,
    top_p=0.9
)

def generate_response(question):
    input_text = f"<|begin_of_text|>User: {question}\nAssistant:"
    outputs = pipe(input_text)
    response = outputs[0]["generated_text"].split("Assistant:")[-1].strip()
    return response

demo = gr.Interface(
    fn=generate_response,
    inputs=gr.Textbox(
        label="‚öñÔ∏è ÿßŸÉÿ™ÿ® ÿßÿ≥ÿ™ŸÅÿ≥ÿßÿ±ŸÉ ÿßŸÑŸÇÿßŸÜŸàŸÜŸä",
        placeholder="ŸÖÿ´ÿßŸÑ: ŸÖÿß ŸáŸä ÿ¥ÿ±Ÿàÿ∑ ŸÅÿ≥ÿÆ ÿπŸÇÿØ ÿßŸÑÿ•Ÿäÿ¨ÿßÿ± ŸÅŸä ÿßŸÑŸÜÿ∏ÿßŸÖ ÿßŸÑÿ≥ÿπŸàÿØŸäÿü"
    ),
    outputs=gr.Textbox(label="üë®‚Äç‚öñÔ∏è ÿßŸÑŸÖÿ≥ÿ™ÿ¥ÿßÿ± ÿßŸÑŸÇÿßŸÜŸàŸÜŸä ÿßŸÑÿ∞ŸÉŸä"),
    title="‚öñÔ∏è ÿßŸÑŸÖÿ≥ÿ™ÿ¥ÿßÿ± ÿßŸÑŸÇÿßŸÜŸàŸÜŸä ÿßŸÑÿ≥ÿπŸàÿØŸä ÿßŸÑÿ∞ŸÉŸä",
    description=(
        "ŸÜÿ∏ÿßŸÖ ÿ∞ŸÉŸä Ÿäÿπÿ™ŸÖÿØ ÿπŸÑŸâ ŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑŸÑÿ∫ÿ© ÿßŸÑŸÖÿ™ŸÇÿØŸÖÿ© ŸÑŸÑÿ•ÿ¨ÿßÿ®ÿ© ÿπŸÑŸâ ÿßŸÑÿ£ÿ≥ÿ¶ŸÑÿ© ÿßŸÑŸÇÿßŸÜŸàŸÜŸäÿ© ÿßŸÑŸÖÿ™ÿπŸÑŸÇÿ© ÿ®ÿ£ŸÜÿ∏ŸÖÿ© "
        "ŸàŸÇŸàÿßŸÜŸäŸÜ ÿßŸÑŸÖŸÖŸÑŸÉÿ© ÿßŸÑÿπÿ±ÿ®Ÿäÿ© ÿßŸÑÿ≥ÿπŸàÿØŸäÿ©. ÿµŸèŸÖŸëŸÖ Ÿáÿ∞ÿß ÿßŸÑŸÖÿ≥ÿßÿπÿØ ŸÑŸäŸÉŸàŸÜ ÿ£ÿØÿßÿ© ŸÖÿ®ÿØÿ¶Ÿäÿ© ÿ™ÿ≥ÿßÿπÿØ ŸÅŸä ÿßŸÑŸÅŸáŸÖ ÿßŸÑŸÇÿßŸÜŸàŸÜŸä "
        "ŸàŸÑÿß ŸäŸèÿπÿØ ÿ®ÿØŸäŸÑÿßŸã ÿπŸÜ ÿßŸÑÿßÿ≥ÿ™ÿ¥ÿßÿ±ÿ© ÿßŸÑŸÇÿßŸÜŸàŸÜŸäÿ© ÿßŸÑÿ±ÿ≥ŸÖŸäÿ©."
    ),
    theme="soft",
    allow_flagging="never"
)

demo.launch(share=True)
```
